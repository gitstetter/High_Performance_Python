```python
# Generators in general
### yield indicates where a value is sent back to the caller, but unlike return, you don't exit the funcion afterwards. 
### Instead the state of the function is remembered

def infinite_sequence():
    num = 0
    while True:
        yield num ###
        num += 1
```

```python
# retrieving next element with next()
gen = infinite_sequence()

print(next(gen))    
print(next(gen))   
print(next(gen))  
```

```python
# Call a generator by a loop
from time import sleep

for i in infinite_sequence():
    print(i, end=' ')
    sleep(0.1)
```

```python
# Comparison list & generator
def fibonacci_list(num_items):
    numbers = []
    a, b = 0, 1
    while len(numbers) < num_items:
        numbers.append(a)
        a, b = b, a+b
    return numbers


def fibonacci_gen(num_items):
    a, b = 0, 1
    while num_items:
        yield a
        a, b = b, a+b
        num_items -= 1
```

```python
fibonacci_list(100)
```

```python
fibonacci_gen(100)
```

```python
%memit len([n for n in fibonacci_gen(100_000) if n % 3 == 0])
```

```python
%memit
sum(1 for n in fibonacci_gen(100_000) if n % 3 == 0)
```

```python
# generators for infinite series
def fibonacci():
    i, j = 0,1
    while True:
        yield j
        i, j = j, i+j
```

```python
# How many fibonacci numbers below 5000 are odd?
def fibonacci_naive():
    i, j = 0, 1
    count=0
    while j <= 5000:
        if j % 2:
            count += 1
        i,j = j, i+j
    return count

def fibonacci_transform():
    count = 0
    for f in fibonacci(): # call generator
        if f > 5000:
            break
        if f % 2:
            count +=1

    return count

fibonacci_naive()
#fibonacci_transform()
```
